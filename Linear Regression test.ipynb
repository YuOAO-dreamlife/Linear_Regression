```python
import numpy as np #設定XY座標範圍
from sklearn import datasets #隨機生成資料
import matplotlib.pyplot as plt #用於繪圖
from sklearn.linear_model import LinearRegression #用於製作線性回歸模組
from sklearn import metrics #計算MSE

x, y = datasets.make_regression(n_samples=150, #樣本數量
                                n_features=1, #特徵數量
                                noise=25) #每筆資料的分散程度

# 調整X的範圍 (range: 0.0 ~ 1.0)
x = np.interp(x, (x.min(), x.max()), (0.0, 1.0))
# 調整Y的範圍 (range: 15.5 ~ 18.5)
y = np.interp(y, (y.min(), y.max()), (15.5, 18.5))

model = LinearRegression() #建立線性回歸模組
model.fit(x, y) #放入測試資料
predict_y = model.predict(x) #根據虛擬資料的x特徵預測可能的y特徵

print("Mean square error(MSE): " + str(metrics.mean_squared_error(y, predict_y)))
print("Mean absolute error: " + str(metrics.mean_absolute_error(y, predict_y)))

plt.scatter(x, y, marker='.', label="True") #隨機生成的資料
plt.scatter(x, predict_y, c='r', marker='.',label="Predicted") #預測資料
plt.legend(loc="upper left") #顯示資料類型的標籤名稱與對應的顏色於左上角
plt.savefig("./result.png")
plt.show() #繪製圖表
```
